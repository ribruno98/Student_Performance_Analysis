# Analysis of student performance

This project analyzes the reading, math, and writing scores of a group of students.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("StudentsPerformance.csv")

subjects = ['Reading', 'Math', 'Writing']
averages = [meanReading, meanMath, meanWriting]
medians = [medianReading, medianMath, medianWriting]
mode = [modeReading.iloc[0], modeMath.iloc[0], modeWriting.iloc[0]]
# Extract the single mode value from each Series

#Reading score
meanReading = df['reading score'].mean()
medianReading = df['reading score'].median()
modeReading = df['reading score'].mode()

#Math score
meanMath = df['math score'].mean()
medianMath = df['math score'].median()
modeMath = df['math score'].mode()

#Writing score
meanWriting = df['writing score'].mean()
medianWriting = df['writing score'].median()
modeWriting = df['writing score'].mode()

font1 = {'family':'serif', 'color':'blue', 'size':20}
font2 = {'family':'serif', 'color':'red', 'size':18}
font3 = {'family':'serif', 'color':'green', 'size':18}

print("Subject averages:")
print("Average Reading:", meanReading)
print("Average Math:", meanMath)
print("Average Writing:", meanWriting)
print(" ")
print("Correlation between subjects:")
print(df[['reading score', 'math score', 'writing score']].corr())

#plot1.bar
plt.figure(figsize=(15, 5))# Add figure and adjust size for better visibility
plt.subplot(1,3,1)
plt.bar(subjects, averages)
plt.title("Average grades per subject", loc='left', fontdict=font1)
plt.ylabel("Averages", fontdict=font2)

#plot2.bar
plt.subplot(1,3,2)
plt.bar(subjects, medians)
plt.title("Median grades per subject", loc="left", fontdict=font1)
plt.ylabel("Medians", fontdict=font2)

#plot3.bar
plt.subplot(1,3,3)
plt.bar(subjects, mode)
plt.title("Mode notes by subject", loc="left", fontdict=font1)
plt.ylabel("Mode", fontdict=font2)
plt.show()

#plot1.hist
plt.figure(figsize=(15, 5)) # Add figure and adjust size for better visibility
plt.subplot(2,3,1)
plt.hist(df['reading score'], bins=10)
plt.title("Distribution of reading scores", loc="left", fontdict=font1)
plt.xlabel("Score", fontdict=font3)
plt.ylabel("Frequency", fontdict=font2)
plt.show()

#plot2.hist
plt.figure(figsize=(15, 5)) # Add figure and adjust size for better visibility
plt.subplot(2,3,2)
plt.hist(df['math score'], bins=10)
plt.title("Distribution of math scores", loc="left", fontdict=font1)
plt.xlabel("Score", fontdict=font3)
plt.ylabel("Frequency", fontdict=font2)
plt.show()

#plot3.hist
plt.figure(figsize=(15, 5)) # Add figure and adjust size for better visibility
plt.subplot(2,3,3)
plt.hist(df['writing score'], bins=10)
plt.title("Distribution of writing scores", loc="left", fontdict=font1)
plt.xlabel("Score", fontdict=font3)
plt.ylabel("Frequency", fontdict=font2)
plt.show()

#plot1.scatter
plt.figure(figsize=(15, 5)) # Add figure and adjust size for better visibility
plt.subplot(2,3,1)
plt.scatter(df['math score'], df['reading score'])
plt.title("Relationship between Math and Reading", fontdict=font1)
plt.xlabel("Math Score", fontdict=font3)
plt.ylabel("Reading Score", fontdict=font2)
plt.show()

#plot2.scatter
plt.figure(figsize=(15, 5)) # Add figure and adjust size for better visibility
plt.subplot(2,3,2)
plt.scatter(df['math score'], df['writing score'])
plt.title("Relationship between Mathematics and Writing", fontdict=font1)
plt.xlabel("Math Score", fontdict=font3)
plt.ylabel("Writing Score", fontdict=font2)
plt.show()

#plor3.scatter
plt.figure(figsize=(15, 5)) # Add figure and adjust size for better visibility
plt.subplot(2,3,3)
plt.scatter(df['reading score'], df['writing score'])
plt.title("Relationship between Reading and Writing", fontdict=font1)
plt.xlabel("Reading Score", fontdict=font3)
plt.ylabel("Writing Score", fontdict=font2)
plt.show()
```

Conclusions:
- The highest average was in reading, followed by writing and math.
- There is a positive correlation between reading and writing scoresmeaning those who do well in one tend to do well in the other.
